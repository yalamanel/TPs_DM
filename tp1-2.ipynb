```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np

def print_separator(title):
    separator = "-" * 40
    print(separator)
    print(title)
    print(separator)


def load_dataset(path):
    df=pd.read_csv(path, header=None)
    df.columns=['attr1','attr2','attr3','attr4','attr5']

    return df


def info_data(data):
    print("nombre de ligne ",data.shape[0])
    print("nombre de colonne ",data.shape[1])
    col=data.columns
    for atr in col:
        print(f"attribut : {atr} type : {data[atr].dtype}")


def calculate_avrage(val):
    avg=sum(val) / len(val)
    return avg


def calculate_median(val):

    sorted_data = sorted(val)    

    n = len(sorted_data)
    middle_index = n // 2

    if n % 2 == 0:

        median = (sorted_data[middle_index - 1] + sorted_data[middle_index]) / 2
    else:

        median = sorted_data[middle_index] 

    return median



def calculate_mode(val):
    element_count = {}   
    for item in val:
        if item in element_count:
            element_count[item] += 1
        else:
            element_count[item] = 1
    max_count = max(element_count.values())
    mode = [item for item, count in element_count.items() if count == max_count]

    return mode


def tandance_centrale(data) : 
    col=data.columns
    for atr in col[:-1]:
        val=data[atr].values.tolist()
        avg=calculate_avrage(val)
        mods=calculate_mode(val)
        med=calculate_median(val)
        print(f"attribut: {atr} : moyenne: {avg:.3f} mode: {mods} median: {med}")


def calculate_quartiles(val):
    sorted_data = sorted(val)
    n = len(sorted_data)
    q0 = sorted_data[0]
    median = calculate_median(sorted_data)
    q1 = calculate_median(sorted_data[:n // 2])
    q3 = calculate_median(sorted_data[n // 2:])
    q4 = sorted_data[-1] 
    print(f"Minimum (Q0): {q0}")
    print(f"Premier Quartile (Q1): {q1}")
    print(f"Median (Q2): {median}")
    print(f"Troisieme  Quartile (Q3): {q3}")
    print(f"Maximum (Q4): {q4}")

    return q0, q1, median, q3, q4


def calculate_all_quartiles(data) : 
    col=data.columns
    for atr in col[:-1]:
        val=data[atr].values.tolist()
        print(f"\nattribut: {atr}\n")
        calculate_quartiles(val)


def calculate_missing_info(val):
    total_values = len(val)
    missing_values = sum(1 for value in val if value is None or (isinstance(value, float) and math.isnan(value)))
    if total_values == 0:
        return (0, 0.0)  

    percentage_missing = (missing_values / total_values) * 100
    return (missing_values, percentage_missing)  


    
def calculate_all_missing(data) : 

    col=data.columns
    for atr in col[:-1]:
        val=data[atr].values.tolist()
        num,per=calculate_missing_info(val)
        print(f"nombre de  valeur manquantes pour l'attribut {atr}: {num}")
        print(f"Pourcentage valeur manquantes pour l'attribut {atr}: {per:.2f}%\n")
```

```python
def generate_box_plot(data, attribute_name, show_outliers=True):
    fig, ax = plt.subplots()
    
    if show_outliers:
        ax.boxplot(data, vert=False, patch_artist=True, labels=[attribute_name])
    else:
        ax.boxplot(data, vert=False, patch_artist=True, labels=[attribute_name], showfliers=False)
    
    plt.title(f"Boîte à Moustache de {attribute_name}")
    plt.grid(True)
    plt.show()
```

```python
attribute_name = "Valeurs"
generate_box_plot(df['attr1'], attribute_name, show_outliers=True)
```
